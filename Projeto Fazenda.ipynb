```python
from ultralytics import YOLOv10

model = YOLOv10(f'{HOME}/train/weights/best.pt')

dataset = sv.DetectionDataset.from_yolo(
    images_directory_path=f'{HOME}/Database:-Projeto-Fazenda---Netcon-1/valid/images',
    annotations_directory_path=f'{HOME}/Database:-Projeto-Fazenda---Netcon-1/valid/labels',
    data_yaml_path=f'{HOME}/Database:-Projeto-Fazenda---Netcon-1/data.yaml'
)

bounding_box_annotator = sv.BoxAnnotator()
label_annotator = sv.LabelAnnotator()
```

```python
import random
import os
from PIL import Image

# Defina os caminhos para as imagens e anotações
images_directory_path = f'{HOME}/drive/MyDrive/Prova de conceito/downy_mildew_images'

# Verifique se o diretório de imagens existe
if not os.path.exists(images_directory_path):
    raise FileNotFoundError(f"O diretório {images_directory_path} não existe.")

# Liste todas as imagens no diretório
image_files = os.listdir(images_directory_path)

# Verifique se há arquivos no diretório
if not image_files:
    raise FileNotFoundError(f"O diretório {images_directory_path} está vazio.")

# Selecione uma imagem aleatória
random_image = random.choice(image_files)
random_image_path = os.path.join(images_directory_path, random_image)

# Verifique se o arquivo de imagem selecionado existe
if not os.path.exists(random_image_path):
    raise FileNotFoundError(f"A imagem {random_image_path} não existe.")

print(f"Imagem aleatória selecionada: {random_image_path}")

# Carregue a imagem usando PIL
image = Image.open(random_image_path)

results = model(source=random_image_path, conf=0.25)[0]
detections = sv.Detections.from_ultralytics(results)

annotated_image = bounding_box_annotator.annotate(
    scene=image, detections=detections
)
annotated_image = label_annotator.annotate(
    scene=annotated_image, detections=detections
)

sv.plot_image(annotated_image)
```
